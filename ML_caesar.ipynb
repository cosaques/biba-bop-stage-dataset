```python
def classifier_morphologie_ligne(row):
    taille = row['taille']
    tour_de_ventre = row['tour_de_ventre_maximal']
    tour_de_poitrine = row['tour_de_poitrine']
    tour_de_taille = row.get('tour_de_taille', tour_de_ventre)  # fallback si non présent
    tour_de_cuisse = row['tour_de_cuisse']

    # Catégorie ventre
    ratio_ventre = tour_de_ventre / taille
    if ratio_ventre < 0.45:
        ventre = "plat"
    elif ratio_ventre < 0.52:
        ventre = "moyen"
    else:
        ventre = "rond"

    # Catégorie torse
    ratio_torse = tour_de_poitrine / taille
    if ratio_torse < 0.48:
        torse = "fin"
    elif ratio_torse < 0.54:
        torse = "moyen"
    else:
        torse = "large"

    # Catégorie cuisses
    ratio_cuisse = tour_de_cuisse / taille
    if ratio_cuisse < 0.30:
        cuisses = "fines"
    elif ratio_cuisse < 0.35:
        cuisses = "moyennes"
    else:
        cuisses = "larges"

    return pd.Series([ventre, torse, cuisses],
                     index=['categorie_ventre', 'categorie_torse', 'categorie_cuisses'])

# Appliquer à ton DataFrame
mesure[['categorie_ventre', 'categorie_torse', 'categorie_cuisses']] = mesure.apply(classifier_morphologie_ligne, axis=1)
```

```python
mesure.head(20)
```

```python
from sklearn.preprocessing import RobustScaler

# Liste des colonnes à scaler
colonnes_a_scaler = ['taille', 'age', 'weight', 
                     'tour_de_ventre_maximal', 'tour_de_poitrine', 'tour_de_cuisse','longueur_du_bras']

# Étape 0 - Instancier le scaler
rb_scaler = RobustScaler()

# Étape 1 - Fit sur les colonnes concernées
rb_scaler.fit(mesure[colonnes_a_scaler])

# Étape 2 - Transformer et remplacer les colonnes d'origine
mesure[colonnes_a_scaler] = rb_scaler.transform(mesure[colonnes_a_scaler])

# Afficher les premières lignes
mesure.head(20)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Colonnes à encoder
colonnes_cat = ['categorie_ventre', 'categorie_torse', 'categorie_cuisses']

# Afficher les valeurs uniques pour vérification
for col in colonnes_cat:
    print(f"Valeurs uniques pour '{col}' :", mesure[col].unique())

# Instancier OneHotEncoder avec drop='if_binary' (garde 1 seule colonne pour les colonnes binaires)
ohe = OneHotEncoder(sparse_output=False, drop='if_binary')

# Fit + transform sur les colonnes sélectionnées
encoded_array = ohe.fit_transform(mesure[colonnes_cat])

# Obtenir les noms des nouvelles colonnes générées
encoded_col_names = ohe.get_feature_names_out(colonnes_cat)

# Créer un DataFrame avec les colonnes encodées
df_encoded = pd.DataFrame(encoded_array, columns=encoded_col_names, index=mesure.index)

# Fusionner avec le DataFrame d’origine
mesure_sc = pd.concat([mesure.drop(columns=colonnes_cat), df_encoded], axis=1)

# Aperçu du DataFrame final
mesure_sc.head(20)
```

```python
mesure_sc.columns
```

```python
from sklearn.model_selection import train_test_split
# split the data into train and test
train_data, test_data = train_test_split(mesure_sc, test_size=0.3)
# Ready X's and y's
X_train = train_data[['taille','age','weight','tour_de_ventre_maximal','tour_de_poitrine','tour_de_cuisse', 'categorie_ventre_moyen',
       'categorie_ventre_plat', 'categorie_ventre_rond', 'categorie_torse_fin',
       'categorie_torse_large', 'categorie_torse_moyen',
       'categorie_cuisses_fines', 'categorie_cuisses_larges',
       'categorie_cuisses_moyennes']]
y_train = train_data['longueur_du_bras']
X_test = test_data[['taille','age','weight','tour_de_ventre_maximal','tour_de_poitrine','tour_de_cuisse', 'categorie_ventre_moyen',
       'categorie_ventre_plat', 'categorie_ventre_rond', 'categorie_torse_fin',
       'categorie_torse_large', 'categorie_torse_moyen',
       'categorie_cuisses_fines', 'categorie_cuisses_larges',
       'categorie_cuisses_moyennes']]
y_test = test_data['longueur_du_bras']
```

```python
model = LinearRegression()
# Train the model on the Training data
model.fit(X_train, y_train)
# Score the model on the Test data
model.score(X_test,y_test)
```
