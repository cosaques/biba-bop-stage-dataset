```python
def classifier_morphologie_ligne(row):
    taille = row['taille']
    tour_de_ventre = row['tour_de_ventre_maximal']
    tour_de_poitrine = row['tour_de_poitrine']
    tour_de_taille = row.get('tour_de_taille', tour_de_ventre)  # fallback si non pr√©sent
    tour_de_cuisse = row['tour_de_cuisse']

    # Cat√©gorie ventre
    ratio_ventre = tour_de_ventre / taille
    if ratio_ventre < 0.45:
        ventre = "plat"
    elif ratio_ventre < 0.52:
        ventre = "moyen"
    else:
        ventre = "rond"

    # Cat√©gorie torse
    ratio_torse = tour_de_poitrine / taille
    if ratio_torse < 0.48:
        torse = "fin"
    elif ratio_torse < 0.54:
        torse = "moyen"
    else:
        torse = "large"

    # Cat√©gorie cuisses
    ratio_cuisse = tour_de_cuisse / taille
    if ratio_cuisse < 0.30:
        cuisses = "fines"
    elif ratio_cuisse < 0.35:
        cuisses = "moyennes"
    else:
        cuisses = "larges"

    return pd.Series([ventre, torse, cuisses],
                     index=['categorie_ventre', 'categorie_torse', 'categorie_cuisses'])

# Appliquer √† ton DataFrame
#mesure[['categorie_ventre', 'categorie_torse', 'categorie_cuisses']] = mesure.apply(classifier_morphologie_ligne, axis=1)

# Appliquer la fonction pour g√©n√©rer les cat√©gories
categories_df = mesure.apply(classifier_morphologie_ligne, axis=1)

# Concat√©ner les anciennes colonnes avec les nouvelles
mesure_cat = pd.concat([mesure, categories_df], axis=1)

# Afficher un aper√ßu
mesure_cat.head()
```

```python
mesure.head(20)
```

```python
from sklearn.preprocessing import RobustScaler

# Liste des colonnes √† scaler
colonnes_a_scaler = ['taille', 'age', 'weight','tour_de_ventre_maximal','tour_de_poitrine','tour_de_cuisse',]

# √âtape 0 - Instancier le scaler
rb_scaler = RobustScaler()

# √âtape 1 - Fit sur les colonnes concern√©es
rb_scaler.fit(mesure_cat[colonnes_a_scaler])

# √âtape 2 - Transformer et remplacer les colonnes d'origine
mesure_cat[colonnes_a_scaler] = rb_scaler.transform(mesure_cat[colonnes_a_scaler])

# Afficher les premi√®res lignes
mesure_cat.head(20)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Colonnes √† encoder
colonnes_cat = ['categorie_ventre', 'categorie_torse', 'categorie_cuisses']

# Afficher les valeurs uniques pour v√©rification
for col in colonnes_cat:
    print(f"Valeurs uniques pour '{col}' :", mesure_cat[col].unique())

# Instancier OneHotEncoder avec drop='if_binary' (garde 1 seule colonne pour les colonnes binaires)
ohe = OneHotEncoder(sparse_output=False, drop='if_binary')

# Fit + transform sur les colonnes s√©lectionn√©es
encoded_array = ohe.fit_transform(mesure_cat[colonnes_cat])

# Obtenir les noms des nouvelles colonnes g√©n√©r√©es
encoded_col_names = ohe.get_feature_names_out(colonnes_cat)

# Cr√©er un DataFrame avec les colonnes encod√©es
df_encoded = pd.DataFrame(encoded_array, columns=encoded_col_names, index=mesure_cat.index)

# Fusionner avec le DataFrame d‚Äôorigine
mesure_sc = pd.concat([mesure_cat.drop(columns=colonnes_cat), df_encoded], axis=1)

# Aper√ßu du DataFrame final
mesure_sc.head(20)
```

```python
mesure_sc.columns
```

```python
from sklearn.model_selection import train_test_split
# split the data into train and test
train_data, test_data = train_test_split(mesure_sc, test_size=0.3)
# Ready X's and y's
X_train = train_data[['taille','age','weight','tour_de_ventre_maximal','tour_de_poitrine','tour_de_cuisse', 'categorie_ventre_moyen',
       'categorie_ventre_plat', 'categorie_ventre_rond', 'categorie_torse_fin',
       'categorie_torse_large', 'categorie_torse_moyen',
       'categorie_cuisses_fines', 'categorie_cuisses_larges',
       'categorie_cuisses_moyennes']]
y_train = train_data['longueur_du_bras']
X_test = test_data[['taille','age','weight','tour_de_ventre_maximal','tour_de_poitrine','tour_de_cuisse', 'categorie_ventre_moyen',
       'categorie_ventre_plat', 'categorie_ventre_rond', 'categorie_torse_fin',
       'categorie_torse_large', 'categorie_torse_moyen',
       'categorie_cuisses_fines', 'categorie_cuisses_larges',
       'categorie_cuisses_moyennes']]
y_test = test_data['longueur_du_bras']
```

```python
model = LinearRegression()
# Train the model on the Training data
model.fit(X_train, y_train)
# Score the model on the Test data
model.score(X_test,y_test)
```

```python
# Discr√©tiser en 2 cat√©gories : 'court' et 'long' selon la moyenne
mesure_sc['longueur_bras_binaire'] = pd.cut(
    x=mesure_sc['longueur_du_bras'],
    bins=[
        mesure_sc['longueur_du_bras'].min() - 1,
        mesure_sc['longueur_du_bras'].mean(),
        mesure_sc['longueur_du_bras'].max() + 1
    ],
    labels=['court', 'long']
)

# V√©rification
mesure_sc[['longueur_du_bras', 'longueur_bras_binaire']].head()
```

```python
mesure_sc.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
# Encode the target
target_encoder = LabelEncoder().fit(mesure_sc['longueur_bras_binaire'])
y = target_encoder.transform(mesure_sc['longueur_bras_binaire'])
# Define the features
X = mesure_sc.drop(columns=['longueur_du_bras', 'longueur_bras_binaire'])
# Instantiate a model
log_reg = LogisticRegression(max_iter=1000)
# Score on multiple folds aka Cross Validation
scores = cross_val_score(log_reg, X, y, cv=10)
scores.mean()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
# Encode the target
target_encoder = LabelEncoder().fit(mesure_sc['longueur_bras_binaire'])
y = target_encoder.transform(mesure_sc['longueur_bras_binaire'])
# Define the features
X = mesure_sc.drop(columns=['longueur_du_bras', 'longueur_bras_binaire'])
# Instantiate a model
log_reg = LogisticRegression(max_iter=1000)
# Score on multiple folds aka Cross Validation
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
scores = cross_val_score(log_reg, X, y, cv=cv)
scores.mean()
```
